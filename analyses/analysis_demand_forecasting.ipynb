```python
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX


# Load the weekly supply of petroleum data in the U.S. (./data/USWeeklyProductSupplied.xlsx) of sheet 'Data 1', cells A17:C1755. Column names are "Date", "n_barrels_total", "n_barrels_motorgasoline". Create the time series for the weekly supply of motor gasoline in the U.S. (n_barrels_motorgasoline) from 1991-02-08 to 2024-05-31.

# Load the data
data = pd.read_excel('./data/USWeeklyProductSupplied.xlsx', sheet_name='Data 1', usecols='A:C', skiprows=16, nrows=1739, index_col=0, header=None, names=['Date', 'n_barrels_total', 'n_barrels_motorgasoline'])
# data.index = data.index.to_period('W')

# Create the time series
data_ts = pd.Series(data['n_barrels_motorgasoline'].values, index=data.index)

data
```

```python
# Define the ARIMA model
model = SARIMAX(data_ts, 
                order=(1, 1, 2), 
                seasonal_order=(0, 0, 2, 52), 
                trend='c')  # 'c' includes a constant term (drift)

# Fit the model
results = model.fit()

# Print the summary of the model fit
print(results.summary())
```

```python
import pandas as pd
import numpy as np

# Define the events and their baseline probabilities
events = [
    'normal',
    'hurricanes',
    'earthquakes',
    'recession',
    'economic boom',
    'oil supply disruptions',
    'peace agreements',
    'renewable energy breakthrough',
    'improved fuel efficiency',
    'new regulations',
    'subsidies for petroleum products',
    'pandemic',
    'large_scale events',
    'severe winter weather',
    'summer travel season',
    'price speculation'
]

# Baseline probabilities for unexpected events per week
event_probabilities = {
    'hurricanes': 0.001,
    'earthquakes': 0.001,
    'recession': 0.003,
    'economic boom': 0.003,
    'oil supply disruptions': 0.002,
    'peace agreements': 0.001,
    'renewable energy breakthrough': 0.001,
    'improved fuel efficiency': 0.001,
    'new regulations': 0.001,
    'subsidies for petroleum products': 0.002,
    'pandemic': 0.0001,
    'large_scale events': 0.0001,
    'severe winter weather': 0.10,
    'summer travel season': 0.75,
    'price speculation': 0.05
}

# Initialize a DataFrame to store probabilities for each month
prob_matrix = pd.DataFrame(index=events, columns=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Adjust probabilities for seasonal variations
for month in prob_matrix.columns:
    for event in events:
        if event in event_probabilities:
            prob_matrix.at[event, month] = event_probabilities[event]
        else:
            prob_matrix.at[event, month] = 0

# Modify for specific events
# Hurricanes: Higher probability in Jun-Nov (Hurricane season)
hurricane_season = ['Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov']
for month in hurricane_season:
    prob_matrix.at['hurricanes', month] = 0.002
# Low probability outside hurricane season
off_season = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Dec']
for month in off_season:
    prob_matrix.at['hurricanes', month] = 0.001

# Severe winter weather: Higher probability in winter months (Dec-Mar)
winter_months = ['Dec', 'Jan', 'Feb', 'Mar']
for month in winter_months:
    prob_matrix.at['severe winter weather', month] = 0.10
# Low probability outside winter months
non_winter_months = ['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov']
for month in non_winter_months:
    prob_matrix.at['severe winter weather', month] = 0.0

# Summer travel season: High probability in summer months (Jun-Aug)
summer_months = ['Jun', 'Jul', 'Aug']
for month in summer_months:
    prob_matrix.at['summer travel season', month] = 0.75
# Lower probability outside summer months
non_summer_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Sep', 'Oct', 'Nov', 'Dec']
for month in non_summer_months:
    prob_matrix.at['summer travel season', month] = 0.0

# Ensure the sum of probabilities for each month is 1
for month in prob_matrix.columns:
    prob_matrix.at['normal', month] = 1 - prob_matrix[month].sum() + prob_matrix.at['normal', month]

prob_matrix.fillna(0.0, inplace=True)

# Rename the rows by replacing spaces with underscores and change the columns to lowercase
prob_matrix.index = prob_matrix.index.str.replace(' ', '_')
prob_matrix.columns = prob_matrix.columns.str.lower()

# Display the probability matrix
prob_matrix
```

```python
import numpy as np

# Define the mean and standard deviation for the impact of each event
impact_distributions = {
    'normal': (0.0, 0.0),  # No impact for normal weeks
    'hurricanes': (-0.075, 0.05),
    'earthquakes': (-0.05, 0.03),
    'recession': (-0.1, 0.05),
    'economic_boom': (0.1, 0.05),
    'oil_supply_disruptions': (-0.15, 0.07),
    'peace_agreements': (0.05, 0.02),
    'renewable_energy_breakthrough': (-0.2, 0.1),
    'improved_fuel_efficiency': (-0.05, 0.02),
    'new_regulations': (-0.07, 0.03),
    'subsidies_for_petroleum_products': (0.08, 0.04),
    'pandemic': (-0.25, 0.1),
    'large_scale_events': (0.05, 0.02),
    'severe_winter_weather': (0.03, 0.015),
    'summer_travel_season': (0.12, 0.05),
    'price_speculation': (0.07, 0.03)
}

# Function to sample a percentage impact based on the distribution
def sample_impact(event):
    mean, std_dev = impact_distributions[event]
    return np.random.normal(mean, std_dev)

# Example of sampling an impact percentage for hurricanes
for event in impact_distributions:
    impact = sample_impact(event)
    print(f"{event.replace('_', ' '):<30}: Impact = {impact:.2%}")
```
